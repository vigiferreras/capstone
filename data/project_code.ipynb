### The Billboard Hot 100s Chart Data Cleanup 

**Reading the data**

```python
hot100_read = pd.read_csv('chart_data.csv')

hot100_df = pd.DataFrame(hot100_read)

hot100_df
```

```python
#checking this dataset
```

```python
hot100_df.info()
```

```python
# week_of is listed as an object
# change this variable to a date type
```

```python
hot100_df['week_of'] = pd.to_datetime(hot100_df['week_of'])

hot100_df.head()
```

```python
# now the week_of column is a datetime data type
hot100_df.info()
```

```python
# next I will alos change the data type of the peak position column
#from object datatype to an integer data type

hot100_df['peak_position'] = hot100_df['peak_position'].astype(int)
hot100_df
```

```python
hot100_df['artist_name'].unique()
```

```python
count_dash = (hot100_df['peak_position'] == '-').sum()
print("Number of rows containing '-':", count_dash)
```

```python
import re

# Define a regular expression pattern to match any artist names
pattern = r'([^,]+)(?: Featuring ([^\,]+))?$'

# Extract all artist names from the 'artist_name' column using the regular expression pattern
all_artists = hot100_df['artist_name'].apply(lambda x: re.findall(pattern, x))

# Create a new column in the DataFrame to store the extracted artist names
hot100_df['all_artists'] = all_artists
hot100_df = pd.DataFrame(hot100_df)
# Display the DataFrame
print(hot100_df)
```
